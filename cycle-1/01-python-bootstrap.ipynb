```python
# Exo B — Docstring et help()
# Écrire `square(n)` qui retourne n**2, AVEC une docstring complète :
#     - description courte sur la 1ère ligne
#     - section "Examples" avec >>> square(3) → 9
# Puis appeler help(square) dans une cellule et regarder ce qui s'affiche.

def square(n) : 
    """
    Fonction qui permet de multiplier n 2x par lui même. Donc revient à faire n*n


    Examples : 
    >>> square(3)
     >>> 9
    """
    return n**2 
help(square)
```

```python
# Exo C — Composition de fonctions
# Écrire 2 fonctions :
#   - `kilos_to_grams(k)` retourne k * 1000
#   - `grams_to_milligrams(g)` retourne g * 1000
# Écrire une 3e `kilos_to_milligrams(k)` qui utilise les 2 précédentes
# en interne (pas faire k * 1_000_000 directement). C'est l'idée de
# composer des fonctions.
# Tester avec kilos_to_milligrams(2) (attendu : 2_000_000).

def grams_to_milligrams(g) : 
    return g * 1000

def kilos_to_grams(k) : 
    return k * 1000

def kilos_to_milligrams(k) : 
    grams = kilos_to_grams(k)
    return grams_to_milligrams(grams)



kilos_to_milligrams(2)
```
