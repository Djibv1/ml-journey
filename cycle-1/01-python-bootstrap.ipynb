```python
# Module 2 Exo C — Composition de fonctions
# Écrire 2 fonctions :
#   - `kilos_to_grams(k)` retourne k * 1000
#   - `grams_to_milligrams(g)` retourne g * 1000
# Écrire une 3e `kilos_to_milligrams(k)` qui utilise les 2 précédentes
# en interne (pas faire k * 1_000_000 directement). C'est l'idée de
# composer des fonctions.
# Tester avec kilos_to_milligrams(2) (attendu : 2_000_000).

def grams_to_milligrams(g) : 
    return g * 1000

def kilos_to_grams(k) : 
    return k * 1000

def kilos_to_milligrams(k) : 
    grams = kilos_to_grams(k)
    return grams_to_milligrams(grams)



kilos_to_milligrams(2)
```

```python
# Module 3 Exo 1 — Rachat de dette : refaire `tier` proprement
# Avec return + elif cette fois.
# tier(20) → "T1", tier(50) → "T2", tier(80) → "T3"

def tier(score):
    if score < 30:
        return "T1"
    elif score <= 70:
        return "T2"
    else:
        return "T3"

# Tests sur les 3 cas de l'énoncé
print(tier(20))   # T1
print(tier(50))   # T2
print(tier(80))   # T3
```

```python
# Module 3 Exo 2 — Mini-fonction utile : mention(score)
# Prend un score sur 100, retourne une mention :
#   - score < 50  → "Insuffisant"
#   - score < 60  → "Passable"
#   - score < 70  → "Assez bien"
#   - score < 80  → "Bien"
#   - score >= 80 → "Très bien"
# Doit gérer un score invalide (négatif ou > 100) → retourner "Invalide"

def mention(score):
    if score < 0 or score > 100:
        return "Invalide"
    elif score < 50:
        return "Insuffisant"
    elif score < 60:
        return "Passable"
    elif score < 70:
        return "Assez bien"
    elif score < 80:
        return "Bien"
    else:
        return "Très bien"

mention(100)
```

```python
# Module 3 Exo 3 — Conditional expression
# Réécrire tier() en UNE SEULE ligne avec des expressions conditionnelles
# imbriquées.

def tier(score) :    
    score = "T1" if score < 30 else ("T2" if score <= 70 else "T3")
    return score
   
    

tier(100)
```

```python
# Module 4 Exo 1 — Indexing & slicing
themes = ['ai-llms', 'dev-code', 'design-ui', 'mindset-perso', 'cuisine', 'philosophie']

# Faire :
# - Print le 1er thème
# - Print le dernier thème SANS utiliser len()
# - Print les 3 premiers thèmes (slicing)
# - Print les 3 derniers thèmes (slicing)
# - Print la liste inversée (slicing avec step)
# - Print 1 thème sur 2

print(themes[0])      # 1er
print(themes[-1])     # dernier (indice négatif, pas besoin de len)
print(themes[:3])     # 3 premiers
print(themes[-3:])    # 3 derniers
print(themes[::-1])   # liste inversée
print(themes[::2])    # 1 sur 2
```

```python
# Exo 2 — Mutations
veille = ['tweet-a', 'tweet-b', 'tweet-c']
# Print veille après chaque opération pour voir l'évolution :
# - Ajouter 'tweet-d' à la fin
# - Insérer 'tweet-z' à l'index 0
# - Retirer 'tweet-b' par sa valeur (pas par index)
# - Trier alphabétiquement la liste EN PLACE (modifie, ne retourne pas)
# - Inverser la liste finale EN PLACE

veille.append("tweet-d")
print(veille)

veille.insert(0, "tweet-z")
print(veille)

veille.remove("tweet-b")   # par valeur, pas pop(index)
print(veille)

veille.sort()              # tri en place (vs sorted qui crée une nouvelle liste)
print(veille)

veille.reverse()           # inverse en place (vs [::-1] qui crée une nouvelle liste)
print(veille)
```

```python
# Exo 4 — Combinaison + tri
mes_notes = ['matrices', 'vecteurs']
new_notes = ['derivee', 'gradient']
# - Combiner les 2 listes avec `+` (créer une 3e liste `toutes_notes`)
# - Faire la même chose avec `extend` sur mes_notes (modifie en place)
# - Print len(mes_notes) — combien d'éléments maintenant ?
# - Utiliser max(mes_notes) pour trouver la note la plus "grande" en ordre alphabétique

toutes_notes = mes_notes + new_notes
print(toutes_notes)         # nouvelle liste, mes_notes inchangée

mes_notes.extend(new_notes)
print(mes_notes)            # mes_notes modifiée en place

print(len(mes_notes))       # 4

print(max(mes_notes))       # 'vecteurs' — ordre lexicographique sur strings
```
